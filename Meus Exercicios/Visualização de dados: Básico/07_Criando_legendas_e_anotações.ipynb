```python
import pandas as pd

import matplotlib.pyplot as plt
from pylab import rcParams

import seaborn as sns

%matplotlib inline
rcParams['figure.figsize'] = 8,4
sns.set_style('whitegrid')
```

```python
caminho = "/home/leonardo/Artificial Intelligence/LinkedIn Learning/arquivos_de_exercicios_python_para_ciencia_dados_basico/dados/mtcars.csv"

carros = pd.read_csv(caminho)
carros.columns = ['nomes', 'mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs',
       'am', 'gear', 'carb']
mpg = carros.loc[:, 'nomes':'mpg']
mpg.head(20)
#mpg.dtypes
```

```python
x = range(1, 10)
y = [1,2,3,4,.5,4,3,2,1]
plt.bar(x, y)
plt.xlabel('nosso título do eixo X')
plt.ylabel('nosso título do eixo Y')
```

```python
import matplotlib.patches as mpatches

plt.plot(mpg)
plt.xlabel('nomes dos carros')
plt.ylabel('milhas/galão')
plt.title('Consumo de gasolina')
plt.xticks(mpg.index, carros.nomes, rotation=45)

legenda = mpatches.Patch(label='mpg')
plt.legend(handles=[legenda])
rcParams['figure.figsize'] = 20,10
plt.savefig('grafico_teste.png')
```
