```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from pylab import rcParams
from scipy.stats.stats import pearsonr

%matplotlib inline
rcParams['figure.figsize'] = 6,3
sns.set_style('whitegrid')

carros = pd.read_csv('mtcars.csv')
carros.columns = ['nomes', 'mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs',
       'am', 'qnt_marchas', 'carb']
```

```python
X = carros[['mpg', 'hp', 'wt', 'qsec']]
sns.pairplot(X)
```

```python
mpg = carros['mpg']
hp = carros['hp']
qsec = carros['qsec']
wt = carros['wt']

coeficiente_pearson, p_valor = pearsonr(mpg, hp)
print('Coeficiente de Pearson {:0.3}'.format(coeficiente_pearson))
```

```python
coeficiente_pearson, p_valor = pearsonr(mpg, wt)
print('Coeficiente de Pearson {:0.3}'.format(coeficiente_pearson))
```

```python
coeficiente_pearson, p_valor = pearsonr(mpg, qsec)
print('Coeficiente de Pearson {:0.3}'.format(coeficiente_pearson))
```
