```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from pylab import rcParams
from scipy.stats import spearmanr
from scipy.stats import chi2_contingency

%matplotlib inline
rcParams['figure.figsize'] = 6, 3
sns.set_style('whitegrid')

carros = pd.read_csv('mtcars.csv')
carros.columns = ['nomes', 'mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs',
       'am', 'qnt_marchas', 'carb']
carros.head()
```

```python
X = carros[['cyl', 'vs', 'am', 'qnt_marchas']]
sns.pairplot(X)
```

```python
cyl = carros['cyl']
vs = carros['vs']
am = carros['am']
qnt_marchas = carros['qnt_marchas']

coeficiente, p_valor = spearmanr(cyl, vs)
print('Correlação de Spearman {:0.3f}'.format(coeficiente))
```

```python
coeficiente, p_valor = spearmanr(cyl, am)
print('Correlação de Spearman {:0.3f}'.format(coeficiente))
```

```python
coeficiente, p_valor = spearmanr(cyl, qnt_marchas)
print('Correlação de Spearman {:0.3f}'.format(coeficiente))
```

```python
tab = pd.crosstab(cyl, am)

chi2, p, dof, exp = chi2_contingency(tab.values)
print('Qui-quadrado: {:0.3} p-valor: {:0.3}'.format(chi2, p))
```

```python
tab = pd.crosstab(cyl, vs)

chi2, p, dof, exp = chi2_contingency(tab.values)
print('Qui-quadrado: {:0.3} p-valor: {:0.3}'.format(chi2, p))
```

```python
tab = pd.crosstab(cyl, qnt_marchas)

chi2, p, dof, exp = chi2_contingency(tab.values)
print('Qui-quadrado: {:0.3} p-valor: {:0.3}'.format(chi2, p))
```
